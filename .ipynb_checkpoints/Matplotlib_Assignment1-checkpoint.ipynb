```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
q1_list = [1,2,3,4,5]
games_rate = {
    'DarkSouls 1':[5,4.1,3.5],
    'DarkSouls 2':[3.5,4.1,3.5],
    'DarkSouls 3':[3.6,4.1,3.5],
    'Bloodborne':[1,3,5],
    'Sekiro':[5,4.1,3.5]
}
q1_dataframe = pd.DataFrame(games_rate,index=['imdb','rotten tomato','Metacritic'])

print('Dataframe content : \n\n',q1_dataframe.head())
plt.hist(q1_dataframe['Bloodborne'].mean(),bins=[1,3,5])
plt.title('Bloodborne Average Rate')
plt.xlabel('rate scale',fontsize=15)
plt.ylabel('Rate',fontsize=15)
plt.show()

plt.plot([2014,2015,2016,2017,2018],[500,550,300,100,50],color = 'blue')
plt.xlabel('Years')
plt.ylabel('Number of sales')
plt.title('Nokia Sales Performance')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
coffee_types = ['Espresso','Tea','Black coffee']
number_of_purchases_morning = [25,10,50]
colors = ['#ff9999','#99ff99','#ffcc39']
plt.pie(number_of_purchases_morning,labels=coffee_types,colors=colors,autopct='%1.1f%%',radius=1,explode=[0.0,0.0,0.0])
plt.title('Morning Beverage Preferences at Dunkin')
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.
